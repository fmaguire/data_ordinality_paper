```python
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
sns.set_style('whitegrid')
sns.set_context('paper')
sns.set_palette('colorblind')
```

```python
dataset_size = 10000
df = {'Data Ordinality': [], '% of Data Erroneous': []}

data = np.zeros(dataset_size)
df['Data Ordinality'].append(0)
df['% of Data Erroneous'].append(0.0)

for order in range(10):
    for ix, value in enumerate(data):
        if np.random.uniform() < 0.05:
            data[ix] += 1
    df['Data Ordinality'].append(order + 1)
    df['% of Data Erroneous'].append((dataset_size - np.sum(data == 0)) / dataset_size * 100)

df = pd.DataFrame(df)
df['Data Ordinality'] = df['Data Ordinality'].astype(int)
```

```python
sns.lineplot(data = df, x='Data Ordinality', y='% of Data Erroneous')
#sns.despine()
plt.ylim(0,100)
plt.xlim(0,10)
plt.xticks(range(0,11))
plt.title('5% Chance of Error')
plt.savefig('data_ord_error.svg')
```
